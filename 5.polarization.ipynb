```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
angles = np.linspace(0, 360, 37)
angles2 = np.linspace(0, 360, 19)
```

```python
diode = np.array([6, 2, 1, 3, 3, 10, 8, 8, 6, 1, -2, -4, 0, 5, 7, 7, 5, 1])/10 + 20
plt.scatter(np.deg2rad(angles2[:-1]), diode)
plt.hlines(diode.mean(), 0, 2*np.pi + 0.3, linestyles='dashed')
plt.xlabel('Angle (deg)')
plt.xticks(np.linspace(0, 2*np.pi, 5), ['0', '90', '180', '270', '360'])
plt.ylabel('Output current (mA)') 
plt.grid(True, which='both')
plt.xlim(0, 2*np.pi + 0.3)
plt.ylim(0, 25)
plt.title('Polarization of light from Diode')
plt.savefig('diode.png', facecolor='w')
plt.show()
```

```python
only_polarizer = np.array([12.4, 12.5, 12.3, 12, 11.7, 11.1, 10.6, 9., 4.8, 0.8, 0.4, 3.7, 8.8, 10.4, 11.2, 11.7, 12 , 12.2, 12.3, 12.4, 12.3, 11.9, 11.5, 10.9, 10.3, 9.2, 4.5, 0.8, 0.4, 3.6, 9.1, 10.5, 11.1, 11.7, 12, 12.3, 12.4])
plt.scatter(np.deg2rad(angles), only_polarizer)
#plt.plot(np.linspace(0, 2*np.pi, 100), max(only_polarizer)*(np.sin(np.linspace(0, 2*np.pi, 100) - np.pi/2)**2))
plt.xlabel('Angle (deg)')
plt.xticks(np.linspace(0, 2*np.pi, 5), ['0', '90', '180', '270', '360'])
plt.ylabel('Output current (mA)') 
plt.grid(True, which='both')
plt.title('Polarization of Laser Light')
plt.xlim(0, 2*np.pi + 0.3)
plt.ylim(0, 15)
plt.savefig('polarizer.png', facecolor='w')
plt.show()
```

```python
malus_law = np.array([2.4, 0.4, 0.075, 1.4, 5.5, 7.8, 9.8, 10.2, 10.3, 10.1, 10.5, 11.2, 10.7, 10.6, 10.1, 9.7, 8.1, 6.5, 2, 0.2, 0.164, 2, 4.3, 9.3, 9.9, 10.4, 10.7, 11.2, 10.8, 11, 10.5, 11.1, 10, 10, 7.7, 5, 1.8])

plt.scatter(np.deg2rad(angles), malus_law)
plt.plot(np.linspace(0, 2*np.pi, 100), max(malus_law)*(np.sin(0.95*np.linspace(0, 2*np.pi, 100) - np.deg2rad(2.5))**2))
plt.xlabel('Angle (deg)')
plt.xticks(np.linspace(0, 2*np.pi, 5), ['0', '90', '180', '270', '360'])
plt.ylabel('Output current (mA)') 
plt.grid(True, which='both')
plt.xlim(0, 2*np.pi + 0.3)
plt.ylim(0, 15)
plt.title('Verification of Malus law')
plt.savefig('malus_law.png', facecolor='w')
plt.show()
```

```python
qwp = np.array([2, 6, 5, 4, 8, 6, 8, 9, 10, 12, 6, 3, -45, 5, 4, 1, 4, 5, 6, 6, 0, 8, 10, 10, 12, 13, 5, 2, 9, 8, 4, 4, 0, 1, -3, -2, 2])/10 + 10

plt.scatter(np.deg2rad(angles), qwp)
plt.hlines(qwp.mean(), 0, 2*np.pi + 0.3, linestyles='dashed')
plt.xlabel('Angle (deg)')
plt.xticks(np.linspace(0, 2*np.pi, 5), ['0', '90', '180', '270', '360'])
plt.ylabel('Output current (mA)') 
plt.grid(True, which='both')
plt.xlim(0, 2*np.pi + 0.3)
plt.ylim(0, 15)
plt.title('Quarter wave plate')
plt.savefig('qwp.png', facecolor='w')
plt.show()
```

```python
hwp = np.array([0.8, 1.3, 5, 9.6, 10.8, 11.4, 12, 12.2, 13, 13, 12.6, 11.1, 11.3, 12.1, 11.9, 10.9, 10, 6.7, 1.1, 1.3, 4.9, 10.4, 11.1, 11.7, 12.3, 12.3, 11.7, 12.8, 13, 12.9, 12.4, 12, 11.5, 10.7, 9.3, 5.2, 1.2])

plt.scatter(np.deg2rad(angles), hwp)
plt.plot(np.linspace(0, 2*np.pi, 100), (hwp.max() - hwp.min())*(np.sin(0.9*np.linspace(0, 2*np.pi, 100)+ np.deg2rad(10))**2) + hwp.min())
plt.xlabel('Angle (deg)')
plt.xticks(np.linspace(0, 2*np.pi, 5), ['0', '90', '180', '270', '360'])
plt.ylabel('Output current (mA)') 
plt.grid(True, which='both')
plt.xlim(0, 2*np.pi + 0.3)
plt.ylim(0, 15)
plt.title('Half wave plate')
plt.savefig('hwp.png', facecolor='w')
plt.show()
```
